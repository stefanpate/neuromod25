```python
from hydra import compose, initialize
from nm.tasks import BehavioralTasks
import matplotlib.pyplot as plt
import numpy as np
```

```python

# context initialization
with initialize(version_base=None, config_path="../config/task", job_name="test"):
    cfg = compose(config_name="wholepop")
```

```python
wholepop_dataset = BehavioralTasks(cfg)
```

```python
for i in range(len(wholepop_dataset)):
    u, nm, target = wholepop_dataset.__getitem__(i)
    t = np.arange(u.shape[0])
    fig, ax = plt.subplots(nrows=2, sharex=True)
    print(f"Task: {i}")
    print(f"NM: {nm[0, 0]}")
    ax[0].plot(t, u)
    ax[0].set_ylabel('Input')
    ax[1].plot(t, target)
    ax[1].set_ylabel('Target')
    ax[1].set_xlabel('t')
    plt.show()
```
